Наприклад, якщо користувач введе "Hello, World!", програма повинна вивести:

8 9

або більш розгорнутий варіант (за бажанням):

4+1+1+1+1=8

```python
inp = input('Input words: ')
dict0 = {'a' : 1,'e' : 1,'i' : 1,'l' : 1,'n' : 1,'o' : 1,'r' : 1,'s' : 1,'t' : 1,'u' : 1,'d' : 2,'g' : 2,'b' : 3,'c' : 3,'m' : 3,'p' : 3,'f' : 4,'h' : 4,'v' : 4,'w' : 4,'y' : 4,'k' : 5,'j' : 8,'x' : 8,'q' : 10,'z' : 10}
inp = inp.lower()
inp = inp.split()
i0 = 0
s = ''
while i0 < len(inp):
    i = 0
    s0 = 0
    while i < len(inp[i0]):
        if isinstance(dict0.get(inp[i0][i]), int) == True:
            s00 = dict0.get(inp[i0][i])
        else:
            s00 = 0
        s0 += s00
        i += 1
    s += str(s0) + ' '
    i0 += 1
print(s)
```
